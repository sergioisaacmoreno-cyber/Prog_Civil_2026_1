## 🏗️ Ejercicio 1: Clasificación de Suelos (Ingeniería Civil)

**Escenario:** En geotecnia, es vital clasificar la consistencia de un suelo fino según su Índice de Plasticidad ($IP$). Un error aquí puede comprometer la estabilidad de una cimentación.

**El Algoritmo:**

1. Leer el valor del Índice de Plasticidad ($IP$).


2. Si $IP$ es menor a 7, el suelo es "Poco plástico".


3. Si $IP$ está entre 7 y 17 (inclusive), el suelo es "Plásticidad media".


4. Si $IP$ es mayor a 17, el suelo es "Altamente plástico".


5. Mostrar el resultado de la clasificación.



**Tu turno (Python):**
# Entrada
IP = float(input("Ingrese el índice de plasticidad (IP): "))

# Proceso
if IP < 7:
    clasificacion = "Poco plástico"
elif 7 <= IP <= 17:
    clasificacion = "Plasticidad media"
else:
    clasificacion = "Altamente plástico"

# Salida
print("Clasificación del suelo:", clasificacion)
