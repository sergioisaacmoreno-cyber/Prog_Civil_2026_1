 # **Dosificación de Mezcla para Columnas**

**Planteamiento**:

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular.

Se sabe que para un concreto estándar de 210 kg/cm2 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

**Requerimientos**:

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna: $$ v = \pi * r ^2 * h $$

. (Sugerencia: usar 3.1416
 o importar math.pi).

4. Calcular la cantidad de sacos:
 $$sacos = v * 9.5 $$

**Prueba del programa**:

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2
):

Volumen ≈0.377 m3
.
Sacos necesarios ≈3.58
 (el estudiante verá aquí la importancia de los tipos decimales).

````python
# Este programa calcula la cantidad de sacos de cemento necesarios para una columna circular.

# ENTRADAS: Sección para solicitar los datos necesarios al usuario.
# La siguiente línea solicita la altura de la columna en metros.
# La función 'input()' muestra el mensaje y espera la entrada del usuario.
# 'float()' convierte la entrada de texto a un número decimal para cálculos.
h = float(input("Ingrese la altura de la columna en metros: "))

# La siguiente línea solicita el diámetro de la columna en metros.
# Al igual que con la altura, 'input()' obtiene el texto y 'float()' lo convierte a número.
d = float(input("Ingrese el diámetro de la columna en metros: "))

# PROCESO: Sección donde se realizan los cálculos.
# Se calcula el radio de la columna dividiendo el diámetro entre 2.
r = d / 2

# Se calcula el volumen de la columna cilíndrica usando la fórmula V = pi * r^2 * h.
# Aquí, 3.1416 es una aproximación de Pi, 'r ** 2' es el radio al cuadrado, y 'h' es la altura.
v = 3.1416 * r ** 2 * h

# Se calcula la cantidad total de sacos de cemento.
# Se multiplica el volumen calculado por 9.5, que es la cantidad de sacos por m³ de mezcla.
sacos = v * 9.5

# SALIDA: Sección para mostrar los resultados al usuario.
# Se imprime el volumen de la columna, formateado a dos decimales (' .2f').
print(f"El volumen de la columna es: {v:.2f} m3")

# Se imprime la cantidad de sacos de cemento necesarios, también formateado a dos decimales.
print(f"La cantidad de sacos necesarios es: {sacos:.2f}")
````
