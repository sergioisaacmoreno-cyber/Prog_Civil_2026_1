> **Relato de partida**  
> **¿Por qué los costeños le decimos “pluma” a la llave del agua?**  
>  
> En el Caribe colombiano, abrir la “pluma” no es solo una acción cotidiana: es una expresión cargada de historia, técnica y cultura. Aunque algunos se sorprendan o incluso se burlen del término, lo cierto es que decir “pluma” en vez de “llave” no es un error, sino una herencia viva que conecta nuestro lenguaje con siglos de evolución hidráulica.  
>  
> Mucho antes de que existieran los grifos modernos, el agua se distribuía mediante sistemas regulados con precisión. En España, por ejemplo, se usaba la **“pluma de agua”** como unidad de medida para calcular el caudal que se entregaba a cada hogar. Aproximadamente, una pluma equivalía a **0.025 litros por segundo**. En algunos lugares, ese flujo se regulaba mediante orificios calibrados y dispositivos llamados **plumeros**, que permitían controlar la salida del agua.  
>  
> Durante la época colonial, muchas de estas prácticas técnicas cruzaron el Atlántico y llegaron a ciudades como Cartagena, Barranquilla y Santa Marta. Con el tiempo, el término “pluma” dejó de ser solamente técnico y pasó al habla cotidiana como sinónimo de la llave del agua.

## 2. Situación problema

Un hogar recibe una cierta cantidad de **plumas de agua**.  
Se sabe que:

- **1 pluma = 0.025 litros por segundo**

Con esta información, el estudiante debe construir un programa que permita calcular:

1. el caudal total en litros por segundo;
2. el consumo diario en litros;
3. el consumo mensual en litros (suponiendo 30 días);
4. una clasificación del consumo:
   - **Bajo**: menos de 1000 litros por día

## 6. Espacio para que el estudiante construya la solución

### Instrucciones
Complete el siguiente bloque de código a partir del algoritmo anterior.

- No copie una solución terminada sin comprenderla.
- Escriba comentarios propios.

- Verifique que los resultados tengan sentido físico.

````python
# ==========================================================
# ACTIVIDAD: MODELAR LA "PLUMA DE AGUA" CON PYTHON
# ==========================================================

# 1. Solicitar al usuario el número de plumas de agua.
# Se utiliza float() para permitir entradas decimales en el número de plumas.
plumas = float(input("Ingrese el número de plumas: "))

# 2. Definir constantes del problema para mejorar la legibilidad y facilitar el mantenimiento.
# CAUDAL_PLUMA: Caudal de una pluma individual en litros por segundo.
CAUDAL_PLUMA = 0.025 # litros/segundo
# SEGUNDOS_DIA: Número total de segundos en un día (24 horas * 60 minutos * 60 segundos).
SEGUNDOS_DIA = 24 * 60 * 60 # segundos
# DIAS_MES: Número de días considerado para el cálculo del consumo mensual (asumiendo un mes de 30 días).
DIAS_MES = 30 # días

# 3. Calcular el caudal total de agua que fluye a través de todas las plumas asignadas.
# El resultado se obtiene en litros por segundo.
caudal_total = plumas * CAUDAL_PLUMA

# 4. Calcular el consumo total de agua en un día.
# Se convierte el caudal por segundo a volumen por día multiplicando por los segundos del día.
consumo_diario = caudal_total * SEGUNDOS_DIA # litros/día

# 5. Calcular el consumo total de agua en un mes.
# Se multiplica el consumo diario por el número de días en el mes.
consumo_mensual = consumo_diario * DIAS_MES # litros/mes

# 6. Clasificar el consumo diario en categorías de "Bajo", "Medio" o "Alto".
# Se usa una estructura if-elif-else para evaluar los rangos de consumo.
nivel_consumo = "" # Inicializar la variable de nivel de consumo
if consumo_diario < 1000:
    nivel_consumo = "Bajo" # Si el consumo diario es menor a 1000 litros
elif 1000 <= consumo_diario <= 3000:
    nivel_consumo = "Medio" # Si el consumo diario está entre 1000 y 3000 litros (inclusive)
else:
    nivel_consumo = "Alto" # Si el consumo diario es mayor a 3000 litros

# 7. Mostrar todos los resultados calculados al usuario de forma clara.
# Se utilizan f-strings para formatear la salida y limitar la precisión de los números flotantes.
print(f"\n--- Resultados del Consumo de Agua ---")
print(f"Número de plumas ingresadas: {plumas}")
print(f"Caudal total estimado: {caudal_total:.3f} litros por segundo")
print(f"Consumo diario estimado: {consumo_diario:.2f} litros")
print(f"Consumo mensual estimado (base 30 días): {consumo_mensual:.2f} litros")
print(f"Clasificación del consumo: {nivel_consumo}")
````

## Explicación Técnica del Código Python: Modelo de Consumo de "Pluma de Agua"

Este documento proporciona una explicación técnica detallada del código Python diseñado para modelar el consumo de agua basado en el concepto histórico de la "pluma de agua".

### 1. Propósito General del Programa

El objetivo principal de este programa es simular y calcular el consumo de agua de un hogar a partir de un número dado de "plumas de agua" asignadas. Utiliza una unidad histórica de caudal (`0.025 litros/segundo` por pluma) para determinar el caudal total, estimar el consumo diario y mensual en litros, y clasificar el nivel de consumo en categorías (Bajo, Medio, Alto).

### 2. Entradas, Procesos y Salidas

*   **Entradas (Inputs):**
    *   `plumas`: Un valor numérico (entero o flotante) que representa la cantidad de plumas de agua asignadas al hogar. Se obtiene a través de la interacción con el usuario mediante la función `input()`.

*   **Procesos (Processing):**
    1.  **Definición de Constantes:** Se establecen valores fijos para el caudal de una pluma (`CAUDAL_PLUMA`), los segundos en un día (`SEGUNDOS_DIA`) y los días en un mes (`DIAS_MES`).
    2.  **Cálculo del Caudal Total:** Multiplica el número de plumas de entrada por el caudal constante de una pluma para obtener el caudal total en litros por segundo.
    3.  **Cálculo del Consumo Diario:** Convierte el caudal total (litros/segundo) a un volumen total consumido en un día (litros/día) multiplicándolo por la cantidad de segundos en un día.
    4.  **Cálculo del Consumo Mensual:** Estima el consumo mensual multiplicando el consumo diario por el número de días en un mes.
    5.  **Clasificación del Consumo:** Aplica una estructura condicional (`if-elif-else`) para categorizar el `consumo_diario` en "Bajo", "Medio" o "Alto" según umbrales predefinidos.

*   **Salidas (Outputs):**
    *   El programa imprime en la consola los siguientes resultados, formateados para una mejor legibilidad:
        *   El número de plumas ingresado.
        *   El caudal total calculado (litros por segundo).
        *   El consumo diario estimado (litros).
        *   El consumo mensual estimado (litros).
        *   La clasificación del consumo (Bajo, Medio, Alto).

### 3. Descripción de Variables Importantes

*   `plumas` (float): Cantidad de unidades de "pluma de agua" asignadas. Es la entrada principal del usuario.
*   `CAUDAL_PLUMA` (float, constante): Valor fijo que representa el caudal de una pluma en litros por segundo (0.025 l/s).
*   `SEGUNDOS_DIA` (int, constante): Valor fijo que representa el total de segundos en un día (86400 segundos).
*   `DIAS_MES` (int, constante): Valor fijo que representa el número de días considerados en un mes para el cálculo (30 días).
*   `caudal_total` (float): El caudal total de agua en litros por segundo para el número de plumas dado.
*   `consumo_diario` (float): El volumen total de agua consumido en un día, en litros.
*   `consumo_mensual` (float): El volumen total de agua consumido en un mes, en litros.
*   `nivel_consumo` (str): Una cadena de texto que indica la clasificación del consumo (Bajo, Medio, Alto).

### 4. Explicación de las Estructuras Usadas

*   **Entrada de Usuario (`input()`):** Se utiliza `input()` para obtener el número de plumas del usuario. La función `float()` se usa para asegurar que la entrada se trate como un número decimal, permitiendo flexibilidad en la cantidad de plumas.
*   **Constantes:** Se emplean variables en mayúsculas (`CAUDAL_PLUMA`, `SEGUNDOS_DIA`, `DIAS_MES`) para indicar que son valores constantes que no deben cambiar durante la ejecución del programa, mejorando la legibilidad y el mantenimiento del código.
*   **Operadores Aritméticos:** Principalmente se usa el operador de multiplicación (`*`) para realizar las conversiones de caudal a consumo y las extensiones de tiempo (segundos a días, días a meses).
*   **Estructura Condicional (`if-elif-else`):** Esta estructura de control se emplea para implementar la lógica de clasificación del consumo. Permite evaluar múltiples condiciones (`consumo_diario < 1000`, `1000 <= consumo_diario <= 3000`) y asignar un valor diferente a `nivel_consumo` según el resultado de estas evaluaciones.
*   **F-strings (`f""`):** Se utilizan para formatear las cadenas de salida, permitiendo insertar fácilmente los valores de las variables directamente dentro del texto y aplicar formatos específicos (e.g., `:.3f` para tres decimales) para una presentación clara de los resultados.
