# Problema de ingeniería

**Calcular el área de un terreno rectangular.**

Este cálculo se usa frecuentemente para:

* estimar **área de cultivo**
* calcular **superficie de construcción**
* dimensionar **parcelas o lotes**

La fórmula es:

[
Área = largo \times ancho
]

---

# 1. Definición del algoritmo

## Entradas

Datos que el usuario debe proporcionar.

* `largo` del terreno (metros)
* `ancho` del terreno (metros)

## Proceso

Operación matemática.

* Multiplicar largo por ancho

[
area = largo * ancho
]

## Salida

Resultado que entrega el programa.

* Área del terreno en **m²**

---

# 2. Pseudocódigo

El pseudocódigo describe la lógica **antes de programar**.

```
INICIO

leer largo
leer ancho

area = largo * ancho

mostrar area

FIN
```

---

# 3. Código en Python

```python
# Programa para calcular el área de un terreno rectangular
# Aplicación básica de ingeniería

# ENTRADAS
largo = float(input("Ingrese el largo del terreno (m): "))
ancho = float(input("Ingrese el ancho del terreno (m): "))

# PROCESO
area = largo * ancho

# SALIDA
print("El área del terreno es:", area, "m²")
```

### ¿Qué es Python?

Desde la perspectiva de un ingeniero, podrías ver a **Python** como una "navaja suiza" digital. Es un lenguaje de programación de **alto nivel**, lo que significa que su sintaxis es muy cercana al lenguaje humano (en inglés), facilitando que te concentres en la lógica del problema (el *qué*) más que en la complejidad de la máquina (el *cómo*).

Para tu carrera, Python es fundamental porque:

* **Versatilidad:** Sirve tanto para un script rápido que calcula la resistencia de una viga, como para una aplicación web compleja en **Streamlit** que gestione datos meteorológicos.

* **Comunidad:** Al ser uno de los lenguajes más usados, casi cualquier problema técnico que enfrentes ya tiene una solución documentada.

### Un ejemplo aplicado a la Ingeniería

Imagina que necesitas calcular el **volumen de agua** necesario para llenar un tanque de almacenamiento cilíndrico en una finca. Antes de escribir código, recuerda que un buen ingeniero define sus variables de forma clara.

Aquí tienes un ejemplo de cómo estructuraríamos esto en una celda de **Google Colab**:

````python
import math

def calcular_volumen_cilindro(radio_m, altura_m):
    """
    Calcula el volumen de un cilindro para almacenamiento de agua.
    Retorna el resultado en metros cúbicos (m3).
    """
    # Usamos la fórmula: V = π * r² * h
    volumen = math.pi * (radio_m ** 2) * altura_m
    return volumen

# Datos de entrada con nombres descriptivos
radio_tanque = 2.5  # metros
altura_tanque = 4.0 # metros

# Llamada a la función
volumen_final = calcular_volumen_cilindro(radio_tanque, altura_tanque)

print(f"El volumen del tanque es: {volumen_final:.2f} m3")
````
